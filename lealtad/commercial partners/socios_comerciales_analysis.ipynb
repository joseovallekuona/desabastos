# Merge lealtad and non lealtad

```python
lealtad_df = pd.read_csv('/Users/joseovalle/Downloads/merged_vendor_analysis_lealtad_v2.csv')
```

```python
not_lealtad_df = pd.read_csv('/Users/joseovalle/Downloads/merged_vendor_analysis_not_lealtad.csv')
```

```python
joined_df = pd.merge(lealtad_df,not_lealtad_df, how = 'outer', on = ['marca_proveedor','from_date','to_date'])
joined_df.drop(columns = ['Unnamed: 0_x','Unnamed: 0_y','_users_y'], inplace = True)
joined_df.drop(columns = ['from_x','to_x'], inplace = True)
joined_df.rename(columns = {
                             'ticket_volume_x':'ticket_volume'
                            ,'vta_amt_x':'vta_amt'
                            ,'total_ticket_sales_x':'total_ticket_sales'
                            ,'_users_x':'_users'
                            ,'ticket_size_x':'ticket_size'
                            ,'from_y':'from'
                            ,'to_y':'to'
                            ,'ticket_volume_y':'ticket_volume_not_promo_lealtad'
                            ,'total_ticket_sales_y':'vta_amt_not_lealtad'
                            ,'ticket_size_y':'ticket_size_not_llealtad'
                            }
                , inplace = True
)

joined_df['week'] = 0
joined_df['year'] = 0
```

```python
joined_df[['marca_proveedor'
          ,'from'
          ,'to'
          ,'ticket_volume'
          ,'vta_amt'
          ,'total_ticket_sales'
          ,'_users'
          ,'ticket_size'
          ,'from_date'
          ,'to_date'
          ,'week'
          ,'year'
          ,'ticket_size_not_llealtad'
          ,'vta_amt_not_lealtad'
          ,'ticket_volume_not_promo_lealtad'
          ,'perfil_premia'
          ,'etiqueta_segmento'
]].to_csv('merged_vendor_analysis_lealtad_vs_no_lealtad_granular.csv')
```
