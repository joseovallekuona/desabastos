```python
dictionary_dfs = {}
for file in files:
    df = pd.read_csv(file,encoding='latin1')
    dictionary_dfs[file] = df


oxxo_dfs = pd.concat(dictionary_dfs.values(), ignore_index= True)
```

```python
oxxo_dfs= oxxo_dfs.dropna(subset=['from'])
oxxo_dfs= oxxo_dfs.dropna(subset=['to'])
```

```python
print(len(oxxo_dfs))
```

```python
oxxo_dfs['from'] = oxxo_dfs['from'].astype(int)
oxxo_dfs['to'] = oxxo_dfs['to'].astype(int)

oxxo_dfs.info()
```

```python
oxxo_dfs.head()
```

```python
oxxo_dfs['from_date'] = pd.to_datetime(oxxo_dfs['from'].astype(str), format='%Y%m%d')
oxxo_dfs['to_date'] = pd.to_datetime(oxxo_dfs['to'].astype(str), format='%Y%m%d')
# oxxo_dfs['row_number'] = oxxo_dfs.sort_values('to_date').reset_index(drop= True).index +1
oxxo_dfs.head()
```

```python
oxxo_dfs.drop(columns = ['Unnamed: 0'], inplace = True)
```
